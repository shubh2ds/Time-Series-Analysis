## PART II: Draw ACF and PACF Plot Charts in Python

```python
#PART II: Draw ACF and PACF Plot Charts in Python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

prices=pd.read_csv("C:/Users/shubh/OneDrive/MyCloud/GIT/Time-Series-Analysis/ACF and PACF/PriceByDays.csv",parse_dates=True,index_col=0)
sm.graphics.tsa.plot_acf(prices.values.squeeze(), lags=15)
sm.graphics.tsa.plot_pacf(prices.values.squeeze(), lags=15)
plt.show()
```

## PART III: From the plot Observations:

```python
#From the plot Observations:

#Autocorrelation(ACF) for MA model 
# => in ACF 2 significant(<5%) lags

#PartialAutocorrelation(PACF) for AR model
# => in PACF 1 (out of 4 -sudden diminishing of correlation after 1 significant lag so rest lags ignored) significant lags

#correlation at lag 0 is ignored.

#to keep the model simple we choose the model having less significant value

#option1:
# => so we choose AR model having 1 significant value. (p, d, q) => (1,0,0)


#option2:
# => combination of AR and MA .(1,0,1)
```
